## Get Heroes link ##

```python
from selenium.webdriver.common.by import By
import re

# get all heroes
heroes_elms = driver.find_elements(By.CLASS_NAME, 'herogridpage_HeroIcon_7szOn')

# get name and link of hero
heroes = [{'name':hero.get_attribute('href').split('/')[-1],'link':hero.get_attribute('href')} for hero in heroes_elms]
```

## Get Heroes Attribute ##

```python
# Function to get all hero attribute
def get_hero_attributes(hero_dict):
    # get link
    driver.get(hero_dict.get('link'))  
    driver.implicitly_wait(5)
    
    # get basic attribute
    get_basic_attribute(hero_dict, 'heropage_AttributeValue_3Gsgg', 'Base')
    get_basic_attribute(hero_dict, 'heropage_AttributeGain_DpX1z', 'Gain')
    
    # Get value element
    get_value_element(hero_dict, 'heropage_ValueElement_3783T')

# Get basic attribute
def get_basic_attribute(hero_dict, class_name, prefix):
    
    # basic attribute
    basis = ['Strength','Agility','Inteligence']
    
    # Get attribute element
    attributes = driver.find_elements(By.CLASS_NAME, class_name)
    
    for i in range(len(attributes)):
        hero_dict[prefix + basis[i]] = float(attributes[i].text)
    
# Get atribute name
def get_attribute_name(elm):
    
    # get elment name
    name = elm.find_element(By.CLASS_NAME,'heropage_SwordIcon_FY7TW').get_attribute('src')
    
    return '_'.join(name.split('/')[-1].split('.')[0].split('_')[1:])

# Get value element
def get_value_element(hero_dict, class_name):
    
    # Get attribute element
    values = driver.find_elements(By.CLASS_NAME, class_name)
    if len(driver.find_elements(By.CLASS_NAME, 'heropage_SwordIcon_FY7TW'))>8:
        for i in range(9):
            hero_dict[get_attribute_name(values[i])] = values[i].text
    else:
        for i in range(8):
            hero_dict[get_attribute_name(values[i])] = values[i].text
        hero_dict['projectile_speed'] = ''
```

```python
# get hero attribute
list(map(get_hero_attributes, heroes))
```

```python
heroes_attributes_df = pd.DataFrame(heroes)
```

### Process data ###

```python
# Change format of percentage
heroes_attributes_df['magic_resist'] = heroes_attributes_df['magic_resist'].apply(lambda x: float(x[:-1])/100)
```

```python
# replace attribute=0 if it's empty
heroes_attributes_df['projectile_speed'] = heroes_attributes_df['projectile_speed'].apply(lambda x: '0' if len(x)==0 else x)
```

```python
# Seperate damge
heroes_attributes_df = pd.concat([heroes_attributes_df, heroes_attributes_df['damage'].apply(lambda x: pd.Series(x.split('-'), index=['damage_start','damage_end']))],axis=1)
heroes_attributes_df = heroes_attributes_df.drop(columns='damage')
```

```python
# Change string to float
attribute_list = ['attack_time','attack_range','projectile_speed','armor','movement_speed','turn_rate']
for attribute in attribute_list:
    heroes_attributes_df[attribute] = heroes_attributes_df[attribute].apply(lambda x: float(x))
```
